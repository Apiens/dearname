```python
import pandas as pd
df = pd.read_csv("../bird_list.csv") #2019국가생물종목록 中 "조류" 복붙한것.
df.head()
```

```python
bird_list= df.values.tolist()
bird_list[0][:5]
```

```python
import requests
from bs4 import BeautifulSoup
```

## Adding wiki thumbnail url

```python
cnt= 0
thumb_url_list = []
for bird_info in bird_list[:]:
    try:
        response = requests.get(f"https://en.wikipedia.org/wiki/{bird_info[2]}")
        soup = BeautifulSoup(response.text)
        thumbnail_tag = soup.select("#mw-content-text > div.mw-parser-output > table.infobox.biota > tbody > tr:nth-child(2) > td > a > img")
        # print(thumbnail_tag)
    
        thumb320 = thumbnail_tag[0]["src"][2:].replace("220px", "320px")
        thumb_url_list.append(thumb320)
    except:
        thumb_url_list.append(f"error on {cnt}")
    print(f"{cnt} {bird_info[2]:<40} {thumb_url_list[-1]}")
    cnt += 1
    
print(thumb_url_list)
```

```python
for i in thumb_url_list:
    if "error" in i:
        print(i)
```
